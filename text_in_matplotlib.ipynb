# Text in matplotlib

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,2,4,2,3]

plt.plot(x,y)

#plt.text(2,3,"java",fontsize=10,style="italic",bbox={"facecolor":"y"})
plt.annotate("python",xy=(2,1),xytext=(3,2.50),arrowprops=dict(facecolor="black",shrink=100))


plt.legend()
plt.show()
```
