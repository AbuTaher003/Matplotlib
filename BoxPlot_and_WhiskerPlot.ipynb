# Box plot and whisker plot

```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["Python"],notch=False,widths=0.2,patch_artist=False, showmeans=True,whis=1.2, sym="go", boxprops={"color":'b'}, capprops={"color":'b'},
           whiskerprops={"color":'b'},flierprops={"markeredgecolor":'r'})

# plt.boxPlot() er details:

# notch = True kore dile box ta konakriti er hobe.
# widths= iccha moto value diye box er width fix kore.
# patch_artist = True kore dile box er vitore color diye fillup hobe.
# showmeans= True kore dile means er upor ekta cursor er moto indicator dekhabe.
# sym="g+" diye upore gol outlier er customize kora jay.
# boxprops, capprops and whiskerprops diye egular customization kora jay. Ei props gula dictionary akare kora hoy.



plt.show()
```

```python
# ekhane amra multiple box banabo.

import matplotlib.pyplot as plt
x = [10,20,30,40,50,60,70]
x1 = [10,20,30,40,50,60,90]

y = [x,x1]

plt.boxplot(y,labels=["Python","C++"],patch_artist=False,showmeans=True,label="BoxPlot")

plt.legend(loc="upper left")

plt.title("Box Plot and Whisker Plot",color = 'r',fontsize=15)
plt.show()
```
