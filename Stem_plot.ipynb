# Stem Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,4,3,6,5]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt='b',) 

# linefmt mane amra line gular design kmn korbo seta. 
# markerfmt mane amra line er upor marker tar design kmn korbo seta. 
# bottom mane amra line gula koto theke suru korbo seta.
# basefmt mane amra base er colour ki dibo seta. 


plt.xlabel("Time",fontsize=15)
plt.ylabel("Speed",fontsize=15)
plt.show()
```
