# Plot stack and area plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
area1 = [2,4,3,4,5]
area2 = [1,3,2,5,4]
area3 = [2,3,5,4,1]

plt.stackplot(x,area1,area2,area3,colors=['r','g','m'],labels=["area1","area2","area3"],baseline="zero")

# colors, labels er jodi multiple value thake tahole egular list akare nite hoy. and ses e s dite hoy. EX: colors
# baseline="zero" means jmn ache tmn e thakbe. baseline "sym" or "wiggle" dile symmetric way te graph show korbe. wiggle ektu alada hoy jodio.


plt.legend()
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("Stack Plot and Area plot")
plt.show()
```
