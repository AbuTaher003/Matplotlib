# Savefig

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,1,3,2,4]

plt.plot(x,y,color='r')

plt.savefig("line") # er kaj holo amra jei folder e achi thik sei folder e graph er image ta save korbe. "line" nam e image ta save hobe. amra er vitor jei nam dibo sei nam diye save hobe.

plt.savefig("line1",dpi=2000,facecolor='g',transparent=False,) 

# ekhane dpi mane hocche dots per inch. jei image er dpi joto besi setar quality toto valo. by default dpi thake 72 or 100.
# facecolor dile image tar frame er color change hoy. eta dekhte folder e giye image ta open koro. 
# transparent=True kore dile facecolor er color tai background e add hobe. mane sada bg remove hobe.
# bbox_inches=("tight") use korle box er seser dike soru hobe i.g tight hobe.
# amra image ta jekono format e save krte pari. pdf, png,jpg etc. For example: "line1.pdf"



plt.show()
```

```python
# another example of savefig and dpi from chatgpt

import matplotlib.pyplot as plt

# ১ম ফিগার: ডিফল্ট dpi
plt.figure(figsize=(5, 5), dpi=72)
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title("Default DPI (72)")
plt.savefig("default_dpi.png")

# ২য় ফিগার: বেশি dpi
plt.figure(figsize=(5, 5), dpi=300)
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title("High DPI (300)")
plt.savefig("high_dpi.png")

plt.show()
```
