# SubPlot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]

plt.subplot(2,2,1)
plt.plot(x,y,color='r')

plt.subplot(2,2,2)
plt.pie([1],colors='r')

plt.subplot(2,2,3)
x1 = [10,20,30,40]
plt.pie(x1)

plt.subplot(2,2,4)
x2 = [1,2,3,4]
y2 = ["java","c","c++","python"]
plt.bar(y2,x2)

plt.show()
```
