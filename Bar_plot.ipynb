 # Bar Plot

```python
import matplotlib.pyplot as plt
x = ["python","C","C++","Java"]
y = [90,70,40,80]
c = ['r','y','g','b']

plt.bar(x,y,color = c,width=0.4,edgecolor='r',linewidth=5,alpha = 1, label = "popularity" )


plt.xlabel("Language",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
plt.title("Popularity of Language",fontsize = 20)
plt.legend()
plt.show()
```

```python
# ekhane amra duita bar plot eksathe korbo but segula overlap korbe. 

x = ["python","C","C++","Java"]
y = [90,70,40,80]
z = [40,50,30,60]
c = ['r','y','g','b']

plt.bar(x,y,color = 'r',label = "popularity" )
plt.bar(x,z,color = 'y',label = "popularity1" )


plt.xlabel("Language",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
plt.title("Popularity of Language",fontsize = 20)
plt.legend()
plt.show()
```

```python
#ekhane amra graph ta horizontally banaisi. tai plt.bar() er sathe h diye plt.barh() korchi. and ekhane barh er vitor width er jaygay height diyechi. 


import matplotlib.pyplot as plt
import numpy as np

x = ["python","C","C++","Java"]
y = [35,60,45,80]
z = [30,50,40,20]

width=0.4
p = np.arange(len(x))
p1 = [j+width for j in p]
plt.xlabel("No.",fontsize=15)
plt.ylabel("Language",fontsize=15)
plt.title("Popularity bar graph",fontsize=20)

plt.barh(p,y,color='r',height=0.4,label = "Popularity" )
plt.barh(p1,z,color='y',height=0.4,label = "Popularity1" )

plt.yticks(p+width/2,x,rotation = 20)

plt.legend()
plt.show() 
```
