# Histogram Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random

x = np.random.randint(10,50,(50)) 
print(x)
```

```python
no = [18, 27, 43, 43, 17, 43, 42, 29, 14, 14, 38, 16, 27, 10, 23, 28, 14, 34, 37, 18, 29, 32, 20, 23, 15, 13, 37, 10, 47, 12, 49, 32, 36, 39, 23, 23, 23, 27, 33, 16, 43, 33, 36, 49, 14, 37, 37, 11, 32, 37]
l = [10,20,30,40,50,60]

plt.hist(no,color='b',edgecolor="r",align="left",log= True,label="Python") 

# hist er vitor: bins=l diye check kore dekho ki hoy. "auto",(0,200) eta use kore dekho ki hoy.ekhetre color,bins kete dite hobe.
# cumulative=-1 er kaj holo boro theke choto te jawa. ar cumulative =1 dile choto theke boro dekhabe.
# bottom =10,20 ja iccha diye dekho. joto diba toto theke suru hobe.
# align="mid","left" diye dekho ki hy.
# histtype ="step/stepfilled/barstacked/bar" egula try kore dekho.
# orientation="horizontal/vertical" diye try kore dekho.
# rwidth = 0.1,0.4,0.5 iccha moto use korle sei onujayi prottekta bar er majhe width kete neya hoy.
# log= True kore dekho ki hoy.

plt.axvline(30,color='y',label="line") # eta diye x axis er upor ekta vertical line diye indicate kora hoy.
plt.legend()
plt.grid(which = "both") # grid akare dekha jabe, which="both" er jonno grid gula box akare hoyeche.
plt.xlabel("Python",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Histogram Plot",fontsize=20,c='r')
plt.show()
```
