# Step plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,1,3,5,4]

plt.step(x,y,color='b',marker="o",ms=10,mfc="r",label="Python")

# plt.step() er details: 

# marker="o" diye marker er type thik kora hoy.
# ms = value diye marker size thik kora hoy.
# mfc = "iccha moto colour" diye marker er color thik kora hy. mfc means marker face color.




plt.legend()
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.grid() # eta diye box box kora hoy.
plt.title("Step Plot",fontsize=15)

plt.show()  
```
