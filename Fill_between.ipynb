# Fill between Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5,6])
y = np.array([1,2,3,4,5,6])

plt.plot(x,y,color='r')
plt.fill_between(x,y,color='g',where=(x>=2)&(x<=4),alpha=.5)

# just x,y dile x axis er nicher jayga ta fillup hoy.
# x=[2,4],y1=2,y2=4 evabe ekta area select kora jay.
# where=(x>=2)&(x<=4) erokom where use korle array gula list hisebe nile hobe na. int value use korte hobe tai egula k numpy array te nite hobe.

plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("Fill between plot",color='r',fontsize=20)
plt.show()
```
